## Предобработка данных


### Замена Nan на 0

Начнем наш процесс перереработки данных с обработки пропусков. Начнем с обработки тех переменных, где пропуски можно легко заменить на 0, так как скорее всего либо пользователь оставил место пустым, либо выгрузка из системы давала nan. При визуальном осмотре таблицы, можно предположить, что это переменные **is_apartment** (так как на Яндекс. Недвижимости можно купить как комнату, так и коммерческую недвижимость, а также апартаменты, в которых нельзя прописаться), **balcony**(Nan- значит балкона нет), **ponds_around3000** (nan- значит нет озер на расстоянии 3 км) и **parks_around3000** (nan- значит нет озер на расстоянии 3 км). Проверим гипотезу по каждому параметру с помощью unique().

```python
display(df['is_apartment'].unique())
```

```python
display(df['balcony'].unique())
```

```python
display(df['ponds_around3000'].unique())
```

```python
display(df['parks_around3000'].unique())
```

Гипотеза подтвердилась - вызовем библиотеку numpy и заменим nan на 0 или False(так как **is_apartment** должны быть **bool**), после этого вызовем info(), чтобы проверить произошли ли изменения, а также сменим тип переменной **is_apartment** на **bool**

```python
import numpy as np
df['is_apartment']=df['is_apartment'].replace(np.nan,False)
df['balcony']=df['balcony'].replace(np.nan,0)
df['ponds_around3000']=df['ponds_around3000'].replace(np.nan,0)
df['parks_around3000']=df['parks_around3000'].replace(np.nan,0)
```

```python
df['is_apartment']=df['is_apartment'].astype('bool') #заменим тип значений на bool
```

```python
df.info() # пропуски удалились, тип переменной для is_apartment стал bool
```

### Удаление пропусков в переменно locality_name

При визуальном анализе мы видим, что в переменной **locality_name** содержаться пропуски, которые мы очно не сможем восстановить из других данных. Посчитаем количество пропусков и если их немного - удалим из выборки, так как данный параметр нам будет крайне важен для анализа в дальнейшем

```python
print('Количесвто пропусков в столбце locality_name:', df['locality_name'].isna().sum())
```

```python
df['locality_name']=df['locality_name'].replace(np.nan,0)
df=df.drop(df[df['locality_name']==0].index)
```

```python
df.info() #пропуски удалены
```

### Заполнение пропусков в переменных airports_nearest и cityCenters_nearest

Продолжаем заполнять пропуски. После того как в переменной **locality_name** не осталось пропусков, мы можем попробовать заменить пропуски в перемннных **airoports_nearest** и **cityCenters_nearest** на медианы, сгруппировав по городам.

```python
df['airports_nearest']= df['airports_nearest'].fillna(df.groupby('locality_name')['airports_nearest'].transform('median'))
```

```python
df['cityCenters_nearest']= df['cityCenters_nearest'].fillna(df.groupby('locality_name')['cityCenters_nearest'].transform('median'))
```

```python
df.info()
```

К сожалению, данная замена не удалась - лишь часть пропусков было заменено. Создадим сводную таблицу и посчитаем значения, чтобы понять, что могло стать причиной того, что  не все пропуски были заменены. Используем одну из переменных- в данном случае 

```python
pivot_airports=df.pivot_table(index='locality_name',values='airports_nearest',aggfunc='count')
display(pivot_airports.sort_values(by='airports_nearest',ascending=True).head(10))
```

Как мы выдим, причиной того, что не все пропуски были заполнены послужило, то что в данном месте отсуствует аэропорт и расстояние до центра города нельзя применить (то есть города/ населенные пункты находятся далеко в Санкт-Петербурга) - заменим значения на ноль. 
Также можно увидеть пролблему с е/ё в переменной **locality_name** - слово поселок написано в два разных варианта - с помощью replace приведем к общему значению. 

```python
df['airports_nearest']=df['airports_nearest'].replace(np.nan,0)
```

```python
df['cityCenters_nearest']=df['cityCenters_nearest'].replace(np.nan,0)
```

```python
df['locality_name']=df['locality_name'].str.replace('поселок','посёлок') #заменили все на одно значение - поселок
```

```python
df.info() #пропуски в этиих ячейки заменены
```

Для того чтобы найти площадь кухни, там где стоит NaN, нужно сделать следующие:
1. Найти "кухонный коэффициент" - отношение медианного значения площади кухник общей площади, аггрегированному по количеству комнат
2. Создать датафрейм и замержить его с основным датафреймом по комнатам.
3. Заполнить пропуски с помощью перемножения общий площади на "кухонный коэффициент"
4. Там где заменить не удалось (так как есть студии и квартиры с открытой планировкой) заменим на 0

```python
group_kitchen_ratio = df.groupby('rooms')['kitchen_area'].median() / df.groupby('rooms')['total_area'].median() #посчитаем коэффиценты
df_kitchen_ratio_total = pd.DataFrame({'kitchen_ratio_total': group_kitchen_ratio}).reset_index()
df = df.merge(df_kitchen_ratio_total, on = 'rooms', how = 'left')#присоеденяем к левой таблице
df['kitchen_area'] = df['kitchen_area'].fillna(value = (df['total_area'] * df['kitchen_ratio_total']))#заполняем перемножением плоащди на кухонной коэффииент
df['kitchen_area']=df['kitchen_area'].replace(np.nan,0)
```

### Добавим столбец с группировкой по этажам

Теперь нужно добавить столбец, который отнесет все этажи в определенную категорию - первый, последний, другой. Для этого необходимо написать функцию, которая пройдет по всему датафрейму и если значение этажа 1 - поставит первый, если значения этажа равно floors_total - последний и другой в любом другом случае.

```python
def my_funcion_floors(row): 
    floor=row['floor']
    total_f=row['floors_total']
    if (floor==total_f) & (total_f!=1):
        return 'последний'
    if floor==1 & total_f==1:
        return 'первый' 
    else:
        return 'другой'

df['floors_grouped'] = df.apply(my_funcion_floors, axis=1)
```

### Считаем корреляции между стоимостью квартиры и другими переменными

В этом пункте нам необходимо выделить корреляции и построить диаграммы распределения для лучшей репрезентации данных. Для того чтобы понять силу корреляции и есть ли связь между переменными обычно выделяют два параметра - силу корреляции и p-value, по которой мы будем принимать или отвергать нулевую гипотезу (если значение p-value меньше 0,05 - значит мы отвергаем нулевую гипотезу и говорим что связь между переменными есть), а для силы этой связи используем коэффицент корреляции.
Нам нужно проанализировать связь между стоимостью квартир и:
<ul>
    <li>площади</li>
    <li>числа комнат</li>
    <li>удаленности от центра</li>
    <li>этажа (первый, последний, любой другой)</li>
    <li>дня недели</li>
    <li>месяца</li>
    <li>года размещения</li>

</ul>

Так как мне хотелось бы посчитать как корреляцию, так и p-value и для облегчение процесса - вызовем библиотеку scipy.stats  и воспользуемся встроенной функции, которая посчитает с помощью уровнения Пирсона.

Начнем с площади. Для анализа этого и других пунктов будем использовать query last_price_k_query (где нет аутлаеров по цене) и создавать новые на основе парамаетров, которыке мы использовали выше (с удалением аутлаеров по тем рамкам, которые мы определили в пункте 4.5)

```python
last_price_k_query_sq.plot(x='last_price_k', y='total_area', kind='scatter',alpha=0.2) 
```

```python
import scipy.stats
corr_sq=scipy.stats.pearsonr(last_price_k_query_sq['total_area'], last_price_k_query_sq['last_price_k'])
display(corr_sq) #cильная корреляция - чем больше площадь, чем больше цена
```

Видим сильную значимую коррелцию - чем больше площадь квартиры, чем выше цена

```python
last_price_k_query_rm.plot(x='rooms', y='last_price_k', kind='scatter',alpha=0.3) 
```

```python
corr_rm=scipy.stats.pearsonr(last_price_k_query_rm['rooms'], last_price_k_query_rm['last_price_k'])
display(corr_rm) 
```

Видим среднюю значимую корреляцию - чем больше комнат, тем больше цена
